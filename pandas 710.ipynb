```python
import pandas as pd
import numpy as np
print('khai báo thư viện thành công')
```

```python
#nhập dữ liệu
file = 'C:/Users/Home/Desktop/code gym 2/FoodPrice_in_Turkey.csv'
data = pd.read_csv(file,header=0, encoding = "ISO-8859-1")
print('*'*20, "Nhập dữ liệu thành công", '*'*20)
display(data)
```

```python
#số cột
print(data.shape[0])
#số hàng
print(data.shape[1])
```

```python
#xác định tên các cột
data.columns
```

```python
#việt hóa tên các cột
#đổi tên chỉ 1 cột xác định
data.rename(columns = {'place':'địa điểm'}, inplace= True)
#đổi tên chỉ 1 vài cột xác định
data.rename(columns = {'place':'địa điểm','product name':'tên sản phẩm'}, inplace= True)

#đổi tên toàn bộ các cột (yêu cầu liệt kê đầy đủ tên cột mới và theo đúng thứ tự)
data.columns = ['địa điểm', 'mã sp', 'tên sp', 'đơn vị','tên đơn vị']
#phải đủ slg tên các cột
data
```

```python
#thêm cột vào bảng đã có
data['cột mới']= np.nan
# data['cột mới']= 0
data
```

```python
#chèn thêm cột chiết khấu vào sau cột giá
data.insert(8,'chiết khấu mới(%)',20)
display(data)
```

```python
#thêm 1 cột trước Gía, giá trị toàn bộ là 50
data.insert(7,'số lượng',50)
display(data)
```

```python
#đổi tên 'tên cột mới' thành tổng và có giá trị là giá*số lượng*(100%-chiết khấu)
# gợi ý: data['A'] = B*C(100-D)/100
data.rename(columns = {'cột mới':'Tổng'}, inplace= True)
data ['Tổng'] = data ['Price'] * data ['số lượng'] * (100-data ['chiết khấu mới(%)'])/100
data
```

```python
data.shape[0]
```

```python
# thêm dòng vào cuối bảng
#sai
# data.loc[data.shape[0]] = ['Sài Gòn','3000', 'Dưa hấu', '5', 'KG', '10','6','2022', '50.000', '0', 'NaN']
#đúng
data.loc[data.shape[0]] = ['Sài Gòn',3000, 'Dưa hấu', 5, 'KG', 10, 10,2022, 50.000,10, 10, np.NaN]
display (data)
```

```python
# thêm dòng vào đầu  bảng
#thêm dòng mới vào đầu bảng
data.loc[-1] = ['Hà Nội',3000, 'Dưa hấu', 5, 'KG', 10, 10,2022, 50.000,10, 10, np.NaN]
#update lại index của bảng
data.index = data.index +1
#sắp xếp lại bảng
data = data.sort_index()
display (data)
```

```python
#Thêm 1 dòng mới vào sau dòng Hà Nội. Với các giá trị lần lượt là 
# 'Đà Nẵng',3002,'Mỳ Quảng',5,'Tô',10,2022,20, 100.000,10,np.NaN
#swap giá trị các dòng
data.loc[-1] = data.loc[0]
data.loc[0] = ['Đà Nẵng',3002,'Mỳ Quảng',5,'Tô',10,2022,20, 100.000,10,10,np.NaN]
#update lại index của bảng

data.index = data.index +1
#sắp xếp lại bảng
data = data.sort_index()
display (data)
```

```python
#cách 2: tạo ra gái trị index trung gian
#chèn thêm dòng ở vị trí bất kì, ví dụ 0.5
data.loc[0.5]= ['Đà Nắng',3002,'Mì quảng',5,'tô',10,9,2022,100.000,20,10,np.NaN]
#sắp xếp lại các hàng theo đúng thứ tự từ nhỏ tới lớn theo số tự nhiên
data = data.sort_index().reset_index(drop=True)
data
```

```python
# xóa cột sử dụng del
data2 = data
display(data2)
#xóa cột 'cột mới' (xóa ngay trên bảng gốc) > nguy hiểm, hạn chế dùng
del data2['chiết khấu(%)']
display(data2)
display(data)
```

```python
# xóa cột sử dụng pop()
data3 = data
print('data3 sau khi tạo')
display(data3)

data4 = data3.pop('chiết khấu mới(%)')
 #kq là cột cần xóa
print('data3 sau khi xóa')   
display(data3)
print('data sau khi xóa')  
display(data)
print('data3 sau khi xóa')  
```

```python
#xóa cột sử dụng hàm drop()
data4 = data
print('data4')
display(data4)
#xóa cột 'cột mới' (ko xóa trực tiếp trên bảng gốc) > nên dùng
data5 = data4.drop('cột mới', axis = 1, inplace=False) #lấy kq trả về
data5 = data4.drop('cột mới', axis = 1, inplace=True) 
#ko trả vê kq và cột bị xóa trực tiếp trên bản gốc

print('data4 sau khi xóa')
display(data4)
print('data5')
display(data5)
print('data gốc')
display(data)
```
